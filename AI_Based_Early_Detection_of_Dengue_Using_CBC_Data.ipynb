```python
import pandas as pd
import numpy as np

# Load the original dataset
# Note: In Google Colab, upload 'CBC Report.csv' to the runtime or mount Google Drive if stored there.
# For this example, assuming it's uploaded to the current directory.
df = pd.read_csv('CBC Report.csv')

# Display original shape
print("Original dataset shape:", df.shape)

# Handle missing values
# Identify numerical and categorical columns
numerical_cols = ['Age', 'Haemoglobin', 'ESR', 'WBC', 'Neutrophil', 'Lymphocyte',
                  'Monocyte', 'Eosinophil', 'Basophil', 'RBC', 'Platelets']
categorical_cols = ['Gender', 'Result']  # Serial and Date are not useful for ML, so we'll drop them later

# Fill missing numerical values with median
for col in numerical_cols:
    if col in df.columns:
        df[col] = df[col].fillna(df[col].median())

# Fill missing categorical with mode (most frequent)
for col in categorical_cols:
    if col in df.columns:
        df[col] = df[col].fillna(df[col].mode()[0])

# Drop rows that are completely empty or have missing critical fields (if any remain)
df = df.dropna(subset=['Result'])  # Ensure label is present

# Optionally, encode categorical for ML readiness
# Map Gender: Male=1, Female=0; Result: Positive=1, Negative=0
if 'Gender' in df.columns:
    df['Gender'] = df['Gender'].map({'Male': 1, 'Female': 0, 'female': 1, 'male': 0})  # Handle case variations
if 'Result' in df.columns:
    df['Result'] = df['Result'].map({'Positive': 1, 'Negative': 0})

# Drop non-feature columns like Serial and Date
columns_to_drop = ['Serial', 'Date']
df = df.drop(columns=[col for col in columns_to_drop if col in df.columns])

# Now, generate 199 noisy copies to make it 200x larger
enlarged_dfs = [df.copy()]  # Start with original

for _ in range(199):
    copy_df = df.copy()
    for col in numerical_cols:
        if col in copy_df.columns:
            # Add Gaussian noise: mean=0, std=10% of column's std
            noise = np.random.normal(0, 0.1 * df[col].std(), size=len(copy_df))
            copy_df[col] += noise
            # Clip to prevent unrealistic negative values (e.g., for counts)
            copy_df[col] = copy_df[col].clip(lower=0)  # Assuming non-negative for most blood metrics
    enlarged_dfs.append(copy_df)

# Concatenate all
enlarged_df = pd.concat(enlarged_dfs, ignore_index=True)

# Display new shape
print("Enlarged dataset shape:", enlarged_df.shape)

# Save the enlarged dataset
enlarged_df.to_csv('enlarged_dengue_dataset_200x.csv', index=False)
print("Enlarged dataset saved as 'enlarged_dengue_dataset_200x.csv'")

# Preview the first few rows
enlarged_df.head()
```
